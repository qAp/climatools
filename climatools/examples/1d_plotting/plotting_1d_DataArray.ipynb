```python
%matplotlib inline

import numpy as np
import xarray as xr

import matplotlib 
import matplotlib.pyplot as plt

import climatools.plot.plot

from IPython import display
```

```python
x = np.linspace(-30, 30, 1000)
y = .02 * (x - 3) * (x - 7) * (x + 20) + 30 * np.sin(x) + 300
y1 = .05 * (x - 5) * (x + 5) * np.cos(x) + 40

da = xr.DataArray(data=y, dims=['x'], coords=[x], name='y')
da1 = xr.DataArray(data=y1, dims=['x'], coords=[x], name='y')
```

```python
# xarray
da.plot()
da1.plot()
```

```python
# invert direction of axes
da.climaviz.plot(xincrease=False, yincrease=False)
da1.climaviz.plot(xincrease=False, yincrease=False)
```

```python
# change axes scale to log (on a positive-valued axis)
da.climaviz.plot(yscale='log')
da1.climaviz.plot(yscale='log')
```

```python
# change axes scale to log (on an axis with negative values)
da.climaviz.plot(xscale='log')
da1.climaviz.plot(xscale='log')
```

```python
# flip the x and y axes
da.climaviz.plot(index_on_yaxis=True)
da1.climaviz.plot(index_on_yaxis=True)
```

```python
# limit the range on the value axis based on 
# the minimum and maximum values on a specified range on the index axis
da.climaviz.plot(index_on_yaxis=True, varlim_from_indexrange=(-30, 0))
da1.climaviz.plot(index_on_yaxis=True)
```

```python
# change the size of the figure
da.climaviz.plot()
da1.climaviz.plot(figsize=(12, 5))
```
